```python
import pandas as pd # importing lib. for analysis
import numpy as np   # importing lib. for numerical computin
```

```python
train_data = pd.read_csv('train.csv')      # fetching the train dataset
train_data
```

```python
train_data.info()           #information about the training dataset
```

```python
train_data['Age'].median()
```

```python
 #filling Nan values for "Age" & "Embarked"
train_data['Age'].fillna(train_data['Age'].median(), inplace= True)
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace= True)
```

```python
#extracting the common names in the "Name" columns
train_data['Name'] = train_data['Name'].str.extract(r',\s*([^\.]*)\s*\.')
train_data['Name'].unique()
```

```python
# replacing less common names with 'rare'
train_data['Name']= train_data['Name'].replace({
    'Don': 'rare',
    'Rev': 'rare',
    'Dr' : 'rare',
    'Mme': 'rare',
    'Sir': 'rare',
    'Col': 'rare',
    'Capt': 'rare',
    'the Countess': 'rare',
    'Jonkheer': 'rare',
    'Lady': 'rare',
    'Major': 'rare',
    'Ms': 'Miss',
    'Mlle':'Mrs'
})
train_data['Name'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder # importing OneHotEncoder for converting the categorical variable into numerical values
encoder = OneHotEncoder(drop= 'first', sparse_output= False)
encoded_sex =  encoder.fit_transform(train_data[['Sex','Name','Embarked']]) # fit and transform the 'Sex','Embarked' & 'Name' column

# creating new dataframe and get new columns
new_cols = encoder.get_feature_names_out(['Sex','Name','Embarked'])
encoded_df = pd.DataFrame(encoded_sex, columns= new_cols, dtype= int) # converting it into dataframe
new_train_data = pd.concat([train_data, encoded_df],axis=1)  # concatinating the dummy variables

# removing 'Sex' & 'Name', 'Embarked' column
new_train_data.drop(['Sex','Name','Embarked'], inplace= True, axis= 1)
new_train_data
```

```python
new_train_data.drop('Cabin', axis=1, inplace= True)
new_train_data.drop('Ticket', axis=1, inplace= True) # this is an important feature
new_train_data['Family_Members'] = new_train_data['SibSp'] + new_train_data['Parch'] + 1
new_train_data['IsAlone'] = (new_train_data['Family_Members'] == 1).astype(int)
```

```python
#creating new feature
# new_train_data['Fare_per_person'] = new_train_data['Fare'] / new_train_data['Family_Members']
```

```python
new_train_data.corr()
```

```python
# dropping all the columns
new_train_data.drop('SibSp', axis=1, inplace= True)
new_train_data.drop('Parch', axis=1, inplace= True)
new_train_data.drop('PassengerId', axis=1, inplace= True)
# new_train_data.drop('Cabin', axis=1, inplace= True)
# new_train_data.drop('Ticket', axis=1, inplace= True)
new_train_data
```
